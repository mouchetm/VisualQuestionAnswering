# In this notebook we present a way to use the model to answer your own questions about your own images

```python
# To be able to use the model to extract the features
from image_model import *
# for the word2vec
import spacy
# For the VQA model and final visualization
from vqa_model import *
```

## Load the VGG16 moddel and build it 

```python
model = VGG_16(weights_path= '../../data/vgg16_weights.h5')
sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd, loss='categorical_crossentropy')
```

## Load w2vec and function to preprocess the questions

### Word2vec chosen is glove (from stanford university)

```python
word_embeddings = spacy.load('en', vectors='en_glove_cc_300_1m_vectors')
```

### Fonction to extract the features from the question 

```python
def get_question_features(question, embedding):
    ''' For a given question, a unicode string, returns the timeseris vector
    with each word (token) transformed into a 300 dimension representation
    calculated using Glove Vector '''
    tokens = embedding(question)
    question_tensor = np.zeros((1, 30, 300))
    for j in xrange(len(tokens)):
            question_tensor[0,j,:] = tokens[j].vector
    return question_tensor
```

## Load and build VQA model

```python
VQA_model_file_name      = '../../experiments/VQA_Demo/models/VQA/VQA_MODEL.json'
VQA_weights_file_name   = '../../experiments/VQA_Demo/models/VQA/VQA_MODEL_WEIGHTS.hdf5'
model_vqa = get_VQA_model(VQA_model_file_name, VQA_weights_file_name)
```

## Label encoder for the outputs of the answers

```python
label_encoder_file_name  = '../../experiments/VQA_Demo/models/VQA/FULL_labelencoder_trainval.pkl_01.npy'
labelencoder = np.load(label_encoder_file_name)
```

# Let's try it on your own image and questions

### Load the image and use vgg16 to extract the 4096 features

```python
path_to_image = "../../data/me.jpg"
```

```python
img = image.load_img(path_to_image, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
Xim = model.predict(x, verbose=1)
```

### Extract the features from the question

```python
# Needs to be a unicode
question = u"Is this men pretty?"
```

```python
Xq = get_question_features(question, word_embeddings)
```

### Score the model to get a probability for the 1000 possible asnwers

```python
y_output = model_vqa.predict([Xq, Xim], verbose=1)
```

### Let's plot the results 

```python
nb_answers = 5
df = retrieve_most_probable_answers(y_output, labelencoder, nb_answers)
df.sort_values(by = 'proba', ascending=True, inplace=True)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.barh(np.arange(df.shape[0]), df.proba, align='center', alpha=0.4)
#ax1.set_yticks(np.arange(df.shape[0]), df.answer)
ax1.set_xlabel('proba')
ax1.set_title(str(question))
ax2.imshow(img)
plt.sca(ax1)
plt.yticks(range(nb_answers), df.answer)
plt.show()
```
