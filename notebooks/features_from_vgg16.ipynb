## Load and compile the model

```python
model = VGG_16(weights_path= '../../data/vgg16_weights.h5')
```

```python
sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(optimizer=sgd, loss='categorical_crossentropy')
```

## Fetch all the images to preprocess then preprocess them

## Preprocess the images and compute the features using the model then save the results as pickles

```python
def create_features_from_img(img_to_load):
    
    out_name = str(get_image_id(img_to_load)) + '.npy'
    
    if os.path.isfile(path_to_save+out_name):
        return 1
    else : 
        img = image.load_img(img_to_load, target_size=(224, 224))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        res = model.predict(x)
        np.save(open(path_to_save+out_name, "wb"), res)
        return 1
```

```python
inputs = path_to_imgs
```

### For speed reason we used parallelization

```python
num_cores = multiprocessing.cpu_count()
```

```python
path_to_save = '../../data/preprocessed_val2014/'
```

```python
%%time
results = Parallel(n_jobs=num_cores)(delayed(create_features_from_img)(i) for i in inputs)
```
